## What is GDP?

Gross domestic product (GDP) at current prices is the GDP at the market value of goods and services produced in a country during a year. In other words, GDP measures the __monetary value of final goods & services produced by a country/state in a given period.__ 

GDP can be broadly divided into goods and services produced by three sectors: 
 - the primary sector (agriculture) 
 - the secondary sector (industry), and 
 - the tertiary sector (services).

It is also known as nominal GDP. More technically, real GDP takes into account the price change that may have occurred due to inflation. This means that the real GDP is nominal GDP adjusted for inflation.

In this assignment, we will use the nominal GDP. Also, we will consider the financial year 2015-16 as the base year, as most of the data required for this exercise is available for the aforementioned period.

```python
from matplotlib import style
from statistics import mean

style.use('ggplot')

# This function is used to compute slope and intercept
# Called from the regression line plots

def find_slope_intercept(xs, ys):    
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    b = mean(ys) - m*mean(xs)
    return m, b
```

<font color = 'blue'> __Handle Missing Values in row for 2015-2016:__ </font> 

For the year 2015-2016, the values are missing for Himachal Pradesh, Maharashtra, Manipur, Mizoram, Nagaland, Punjab, Rajasthan, Tripura, Andaman & Nicobar Islands. Since we have removed the Union Territories, there are 8 states where the values need to be imputed. 

These missing values can be imputed in various ways: 
 - take the __average__ of the previous financial years; however mean is not a right measure and it is not recommended. 
 - assume the __same percentage growth__ as in the previous financial year; while charting with growth rates, this will result in a flat line.
 - __compute the % growth__ based on GSDP of previous years: this is a better alternative. 

With the third approach we will compute the missing values as follows:

 - Compute (GSDP in 2014-2015 / GSDP in 2011-2012) 
 - Raise it to the power of (1 / number of periods). Here the number of periods = 4 (2011-2012, 2012-2013, 2013-2014, 2014-2015)
 - Deduct 1 from the resultant value.
 - The result will be a decimal that need to be expressed as a percentage by multiplying it with 100 (2 decimal precision).
 - The computed percentage value can be used to compute the GSDP for 2015-2016 from GSDP of 2014-2015.
 
The formula is explained in https://m.wikihow.com/Calculate-an-Annual-Percentage-Growth-Rate

```python
def compute_growth_rate(gdp_final, gdp_start, periods):
    return round((((gdp_final / gdp_start) ** (1 / periods)) - 1) * 100, 2)

for state in states:
    if data[state].isnull()[4] == True:
        growth_rate = compute_growth_rate(data[state][3], data[state][0], 4)
        gsdp = round(data[state][3] * (100 + growth_rate) / 100, 0)
        data.loc[4, state] = gsdp
        data.loc[9, state] = growth_rate
```

### Task 3: Contribution of different Sectors

The next task is to plot the percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states.The GDP is captured for each state under various items produced by 3 sectors: (a) the primary sector (agriculture) (b) the secondary sector (industry), and (c) the tertiary sector (services). The csv files contain summary rows also, that gives the total of primary, secondary and tertiary. Total GSDP is also available for the state. We need to build a dataframe with these 4 rows for each state to make this computation.

```python
# to get the relevant data from our dataset

required_rows = ['Primary', 'Secondary', 'Tertiary', 'Gross State Domestic Product']

sector_data = data1b.loc[data1b['Item'].isin(required_rows)] 

sector_data = sector_data.drop(['Serial'], axis = 1) # Drop the serial number field
```

```python
# the data set need to be converted into a pivot (rows to columns)

sector_pivot = pd.DataFrame(sector_data.pivot_table('Value', 'State', 'Item').to_records())

sector_pivot.rename(columns= {"Gross State Domestic Product": "GSDP"}, inplace=True)

sector_pivot.sort_values(['GSDP'], inplace = True) # sort based on the GSDP value

sector_pivot['Others'] = sector_pivot['GSDP'] - sector_pivot['Primary'] - sector_pivot['Secondary'] - sector_pivot['Tertiary']
```

```python
sector_plot = sector_pivot[['State', 'Primary', 'Secondary', 'Tertiary', 'Others']]

sector_plot.plot(x='State', kind='bar', figsize=(15,7), stacked=True) 

plt.xlabel('States Of India')
plt.ylabel('GSDP - divided as Primary, Secondary, Tertiary')

plt.title('GDP Of States for the year 2014-15')
```

After this step we have a dataframe with 27 records, one for each state with the contribution of primary, secondary and tertiary sectors and others (adding these 4 values will give the GSDP). We need to express this as percentage while plotting. A stacked bar with the percentage will express the percentage contribution of primary, secondary, tertiary in terms of the total.

```python
# to plot the percentage, find the percentage of primary, secondary and tertiary

sector_pivot['Primary_Percentage'] = round((sector_pivot['Primary'] / sector_pivot['GSDP']) * 100, 2)
sector_pivot['Secondary_Percentage'] = round((sector_pivot['Secondary'] / sector_pivot['GSDP']) * 100, 2)
sector_pivot['Tertiary_Percentage'] = round((sector_pivot['Tertiary'] / sector_pivot['GSDP']) * 100, 2)
sector_pivot['Difference'] = 100 - sector_pivot['Primary_Percentage'] - sector_pivot['Secondary_Percentage'] - sector_pivot['Tertiary_Percentage']
```

```python
sector_plot_percentage = sector_pivot.drop(['GSDP', 'Primary', 'Secondary', 'Tertiary', 'Others'], axis = 1)
```

```python
sector_plot_percentage.plot(x='State', kind='bar', figsize=(15,7), stacked=True) 

plt.xlabel('States Of India')
plt.ylabel('GSDP Percentage contributions from Primary, Secondary, Tertiary')

plt.title('GDP Of States for the year 2014-15')
```

__Which plot will you use here? Why?__ Here I used a stacked bar to express the percentage contribution of each of the sector to the total GSDP, because it clearly shows how the three sectors add up and contribute to the overall GSDP.

__Why is (Primary + Secondary + Tertiary) not equal to total GDP?__ 

The sum of Primary, Secondary and Tertiary will add up to total GSVA. However, there is a tax on this and subsidies. 

GSVA = Primary + Secondary + Tertiary

GSDP = GSVA + Taxes on the Products - Subsidies on the Products.
