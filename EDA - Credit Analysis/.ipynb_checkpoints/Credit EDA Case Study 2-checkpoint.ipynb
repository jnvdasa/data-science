### 7C. Derived Features

We can extract the following information about the client from previous_applications:

 - REJECTED: How many loans were rejected?
 - APPROVED: How many loans were approved?
 - CANCELED: How many loans were canceled?
 - UNUSED: How many loans were unused?
 - TOTAL_APPLICATIONS: How many applications in the past?
 - TOTAL_LOAN_ASKED: What is the total amount the client asked for?
 - TOTAL_LOAN_APPROVED: What is the total amount given as loan to the client?
 - CONSUMER: How many consumer loans?
 - CASH: How many cash loans?
 - REVOLVING: How many revolving loans?
 - DELAY in Payment (LAST_DUE_DATE < DAYS_TERMINATION)
 - Early Payments (LAST_DUE_DATE > DAYS_TERMINATION)
 
These derived fields are then merged with the application data.

```python
features = ['SK_ID_CURR', 'NAME_CONTRACT_TYPE', 'NAME_CONTRACT_STATUS', 
            'NAME_CLIENT_TYPE', 'AMT_APPLICATION', 'AMT_CREDIT', 'CNT_PAYMENT', 
            'DAYS_FIRST_DRAWING', 'DAYS_LAST_DUE', 'DAYS_TERMINATION']

df = df_previous_application[features] # Choosing specific features required for further analysis
```

```python
df['COUNT'] = 1 # Adding a column COUNT
```

```python
# Creating aggregated columns based on contract status (approved, refused, cancelled, unused)

table1 = pd.DataFrame(df.pivot_table(index = 'SK_ID_CURR', 
                                     values = 'COUNT', 
                                     columns = 'NAME_CONTRACT_STATUS', 
                                     aggfunc = np.sum).to_records())
```

```python
# Creating aggregated columns based on contract type (cash, consumer, revolving)

table2 = pd.DataFrame(df.pivot_table(index = 'SK_ID_CURR', 
                                     values = 'COUNT', 
                                     columns = 'NAME_CONTRACT_TYPE', 
                                     aggfunc = np.sum).to_records())
```

```python
# Creating aggregated columns based on total amount applied for, total amount credited and number of loans for the client

table3 = pd.DataFrame(df.pivot_table(index = 'SK_ID_CURR', 
                                     values = ['AMT_APPLICATION', 'AMT_CREDIT', 'COUNT'],
                                     aggfunc = np.sum).to_records())
```

```python
# Delayed Payment and Early Payment
```

```python
df = df_previous_application[['SK_ID_CURR', 'DAYS_LAST_DUE', 'DAYS_TERMINATION']]
```

```python
# 365243 is an invalid value - to be imputed with null values so that it does not affect the analysis

df.loc[df['DAYS_LAST_DUE']    == 365243, 'DAYS_LAST_DUE'   ] = np.nan
df.loc[df['DAYS_TERMINATION'] == 365243, 'DAYS_TERMINATION'] = np.nan
```

```python
df['PAY_DELAY'] = round((df['DAYS_LAST_DUE'] - df['DAYS_TERMINATION']) / 365, 2)
```

```python
df.loc[df['PAY_DELAY'] < 0, 'DELAY_CNT'] = 1
df.loc[df['PAY_DELAY'] > 0, 'EARLY_CNT'] = 1
```

```python
df.fillna(0, inplace = True)
```

```python
table4 = pd.DataFrame(df.pivot_table(index = 'SK_ID_CURR', 
                                     values = ['DELAY_CNT', 'EARLY_CNT'],
                                     aggfunc = np.sum).to_records())
```

### 7D. Merging the Datasets

```python
merged  = pd.merge(table1, table2, on = 'SK_ID_CURR', how = 'inner')
```

```python
merged = pd.merge(merged, table3, on = 'SK_ID_CURR', how = 'inner')
```

```python
merged = pd.merge(merged, table4, on = 'SK_ID_CURR', how = 'inner')
```

Surprisingly, the unemployed category is having a higher credit compared to others (and no outliers for this category).  In case of students there are no cases reported to have payment difficulty. Median value for credit amount is in the range of 500000 (except for unemployed)

The number of clients with academic degree is very less (164 records) in the dataset. 71% of the applicants have only secondary education. However, when we compute the percentile, we see that with whatever sample data we have the median value is higher for the clients with academic degree. The credit amount progressively increases from lower secondary, to secondary, to incomplete higher, to higher education and to academic degree. This variable has direct impact on how much credit the client is eligible for.

The credit median is comparatively higher for the married people and lesser for widows. Those in Rented Apartments or staying with parents has a lower credit limit. Managers have a higher credit limit followed by realty agents & accountants.  Credit for HR is better than that for IT.

## Step 10. Identifying the Driver Variables

```python
def plot_target_segment(column):
    
    df = df_merged[['TARGET', column]]
    
    df['TGTVAL'] = df['TARGET'].apply(str)
    
    df_pivot = pd.DataFrame(df.pivot_table(index = column, 
                                           columns = 'TGTVAL', 
                                           values = 'TARGET', 
                                           aggfunc = 'count').to_records())
    df_pivot.fillna(0, inplace = True)
    
    df_pivot['%OF1'] = round((df_pivot['1'] / (df_pivot['0'] + df_pivot['1'])) * 100, 2)
    df_pivot['%OF0'] = round((df_pivot['0'] / (df_pivot['0'] + df_pivot['1'])) * 100, 2)
    
    df_pivot.drop(['0', '1'], axis = 1, inplace = True)
    df_pivot.sort_values(by = '%OF1', ascending = False, inplace = True)
    
    plt.figure(figsize=(15, 10))
    df_pivot.plot(x = column, kind = 'bar', figsize=(15,7), legend = False, stacked = True)
    
    plt.title(column + ' vs. TARGET')
    plt.tick_params(labelsize=12)
```

```python
segments = ['NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 
            'NAME_HOUSING_TYPE', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'OCCUPATION_TYPE',
            'NAME_CONTRACT_TYPE', 'AMT_INCOME_RANGE', 'AMT_CREDIT_RANGE', 'AGE_GROUP'] 
```

```python
for column in segments: plot_target_segment(column)
```

```python
quantitative = ['CNT_CHILDREN', 'CNT_FAM_MEMBERS', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 
                'TOTAL_APPLIED_PAST','TOTAL_CREDIT_PAST', 'DELAY_CNT', 'EARLY_CNT']
```

 - 7-8% irrespective of whether the client owns a car / realty. Not a Predictor.
 - The Middle Age and Youth have more difficulties compared to the Senior & Elderly.
 - Payment difficulty is more probable with those who stay in rented apartment or with parents.
 - Low-skill Laborers, Drivers, Waiters / Barmen Staff, Security Staff, Laborers, Cooking Staff have more than 10% defaulters.
 - Applicant age and experience has a greater kernel density distribution for defaulters.
 - Delays in the past payment is another strong driver.
